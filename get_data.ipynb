```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 读取 CSV 文件
file_path = 'train.csv'
data = pd.read_csv(file_path)

# 选择前几行进行可视化
num_images = 5  # 要显示的图像数量

for i in range(num_images):
    # 提取第 i 行的数据
    image_data = data.iloc[i, 1:].values  # 从第1列开始是像素数据，跳过标签列
    image_data = image_data.reshape(28, 28)  # 将784个像素值重塑为28x28的图像

    # 显示图像
    plt.figure()
    plt.imshow(image_data, cmap='gray')
    plt.title(f'Label: {data.iloc[i, 0]}')  # 图像的标签
    plt.axis('off')  # 隐藏坐标轴
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取 CSV 文件
file_path = 'train.csv'
data = pd.read_csv(file_path)

# 统计每个类别的样本数量
label_counts = data['label'].value_counts().sort_index()

# 打印每个类别的样本数量
print(label_counts)

# 可视化各类别的分布
plt.figure(figsize=(10, 6))
label_counts.plot(kind='bar', color='skyblue')
plt.title('Distribution of Each Digit in the Dataset')
plt.xlabel('Digit Label')
plt.ylabel('Number of Samples')
plt.xticks(rotation=0)
plt.show()
```
